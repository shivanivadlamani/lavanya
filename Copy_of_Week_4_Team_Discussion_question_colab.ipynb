**Write a Python program to find the indices of the closest pair from a list of numbers.**

**Input:**
[1, 7, 9, 2, 10]

**Output:**
[0, 3]

```python
# Write your query here
l = eval(input())
ls = sorted(l)
min_dis = ls[1]-ls[0]
min_in = (0,l.index(ls[1]))

for i in range(1,len(ls)-1):
  diff = ls[i+1] - ls[i]
  if diff < min_dis:
    min_dis = diff
    min_in = (l.index(ls[i]),l.index(ls[i+1]))
print(min_in)
```

### **Python-2**

**Write a Python program to find the closest palindrome from a given string.**

**Example 1:**

**Input:**
cat

**Output:**
cac

Example 2:

**Input:**
madan

**Output:**
madam

```python
# Write your query here
def pal(word):
  mid = len(word)//2

  if len(word)%2 == 0:
    first = word[:mid]
    last = first[::-1]

  else:
    first = word[:mid+1]
    last = first[:-1][::-1]

  palindrome = first + last
  return palindrome

word = input()
print(pal(word))
```

### **Python-3**

**Write a Python program to find the index of the matching parentheses for each character in a given string.**

**Input:**
()(())

**Output:**
[1, 0, 5, 4, 3, 2]

```python
# Write your query here
def find_matching_parentheses(s):
  # create a list to store the positions of all "("
  stack = []
  # create a list to store the output
  output = [0]*len(s)

  for i in range(len(s)):
    if s[i] == "(":
      # add the current position to the stack
      stack.append(i)
    else:
      # find the position of the matching "(" in the stack
      matching_index = stack[-1]
      # remove the last item from the stack by slicing it out
      stack = stack[:-1]

      # store the matching parentheses positions
      output[i] = matching_index
      output[matching_index] = i
  return output

input_string = "()(())"
print(find_matching_parentheses(input_string))
```

### **Python-4**

**Write a Python program to create a list whose ith element is the maximum of the first i elements from a input list.**

**Input:**
[0, -1, 3, 8, 5, 9, 8, 14, 2, 4, 3, -10, 10, 17, 41, 22, -4, -4, -15, 0]

**Output:**
[0, 0, 3, 8, 8, 9, 9, 14, 14, 14, 14, 14, 14, 17, 41, 41, 41, 41, 41, 41]

```python
# Write your query here

def max_till_i(input_list):
    max_list = []
    for i in range(1, len(input_list) + 1):
        max_list.append(max(input_list[:i]))
    return max_list

input_list = [0, -1, 3, 8, 5, 9, 8, 14, 2, 4, 3, -10, 10, 17, 41, 22, -4, -4, -15, 0]
print(max_till_i(input_list))
```

### **Python-5**

**Write a Python Program to Determine Whether one String is a Rotation of Another**

**Input:**
Given first string = "almacoding"

Given second string = "codingalma"

**Output:**
The given second string is the rotation of the given second string

```python
# Write your query here
# Define function that takes two strings as input
def is_rotation(s1, s2):
    # Join the two copies of the first string
    double_s1 = s1 + s1
    # Check if the second string exists in the joined string
    if s2 in double_s1:
        result = True   # s2 is a rotation of s1
    else:
        result = False  # s2 is not a rotation of s1
    return result

# Use the function on your input
s1 = "almacoding"
s2 = "codingalma"

# Print the result
if is_rotation(s1, s2):
    print("The given second string is the rotation of the given first string")
else:
    print("The strings not rotations of each other")
```
